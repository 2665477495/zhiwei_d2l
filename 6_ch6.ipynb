# 1.1 训练

```python
%matplotlib inline
import torch
from torch import nn
import matplotlib.pyplot as plt
```

```python
T = 1000  # 生成数据集的总数
time = torch.arange(1, T + 1, dtype=torch.float32)  # 时间索引
x = torch.sin(0.01 * time) + torch.normal(0, 0.2, (T,))  # 生成序列
plt.plot(time.numpy(), x.numpy(), label='x')
```
